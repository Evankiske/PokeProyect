```python
#Importamos las librerias a Usar
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
#Creamos una lista vacia para empezar a hacer web-scrapping
#Creamos un loop para leer y traer la tabla deseada de 809 paginas

lista = []

for i in range(1, 810):
    url = f'https://www.serebii.net/pokedex-sm/{str(i).rjust(3,"0")}.shtml'

    html = requests.get(url).content
    tabla = pd.read_html(html)
    links = BeautifulSoup(html, 'html5lib')
    
    tables = links.find_all('table')
    variable = [table for table in tables if table.find('b') and table.find('b').text == 'Stats']
    stats = pd.read_html(str(variable[0]))[0]
    lista.append(stats)
```

```python
print(len(lista))
```

```python
#Creamos Una segunda lista para no modificar la primer por si algo falla
lista2 = lista.copy()
```

```python
#hacemos concatenacion a la lista original 
datatrue = pd.concat(lista)
```

```python
#Salvamos en. un DataFram Suciocio para futuros usos
datatrue.to_csv('PokeStatsNotClean', sep='\t', index=False)
```

```python
#Convertimos la nueva lista en un DataFrame
data = pd.concat(lista2)
```

```python
print(type(data))
```

```python
#Eliminamos la columna extra que no necesitamos
data = data.drop(8, axis = 1)
```
