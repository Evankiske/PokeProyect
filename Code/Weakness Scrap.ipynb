```python
#Importamos las librerias a usar
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
#Leemos el csv
df = pd.read_csv('Pokemon.csv')
```

```python
#Creamos una lista con los nombres para iniciar el scrap
nam = df.Name
```

```python
#Empesamos el scrapeo, usando una lista para meter los resultado
lista = []

for i in nam:
    print(i)
    #url = f'https://bulbapedia.bulbagarden.net/wiki/{}_(Pokémon).format(e)'
    #url = f'https://bulbapedia.bulbagarden.net/wiki/Flabébé_(Pokémon)'
    html = requests.get('https://bulbapedia.bulbagarden.net/wiki/{}_(Pokémon)'.format(i)).content
    tabla = pd.read_html(html)
    links = BeautifulSoup(html, 'html5lib')

    tables = links.find_all('table')
    variable = [table for table in tables if table.find('b') and table.find('b').text == 'Under normal battle conditions in Generation VII, this Pokémon is:']
    stats = pd.read_html(str(variable[0]))[0]
    lista.append(stats)
```

```python
#Conatenamos los datos para tenerlos en un solo DF
datatrue = pd.concat(lista)
```

```python
#Salvamos en un csv los datos para futura manipulacion
datatrue.to_csv('PokeDamageNoClean.csv', sep='\t', index=False)
```

```python
#Creamos una copia de los datos para manipular
data = datatrue.copy()
```

```python
#Vemos el contenido del Data Set
data.head()
```

```python
#Renombramos columnas para su facil manipulacion
data.rename(columns={0:'Damaged', 1:'X'}, 
                 inplace=True)
```

```python
#Revisamos los cambios realizados
data.head()
```

```python
#Eliminam os las columnas con elementos Vacios
data = data.dropna()
```

```python
#Revisamos el nuevo data set
data.head(20)
```

```python
#Separamos los datos para salvar futuros csv para su facil manipulacion, usamos el tipo de
#debilidad para crear nuestros nuevos Data Sets usando palabras clave
normally = data[data['Damaged'].str.contains("Damagednormally")]
```

```python
weak = data[data['Damaged'].str.contains("Weak to")]
```

```python
inmune = data[data['Damaged'].str.contains("Immune to")]
```

```python
resist = data[data['Damaged'].str.contains("Resistant to")]
```

```python
#Salvamos los nuevos Data Sets para futura manipulacion y agregarlos al data set con el que se trabajara
normally.to_csv('PokeElN.csv', sep='\t', index=False)
weak.to_csv('PokeElW.csv', sep='\t', index=False)
inmune.to_csv('PokeElI.csv', sep='\t', index=False)
resist.to_csv('PokeElR.csv', sep='\t', index=False)
```
